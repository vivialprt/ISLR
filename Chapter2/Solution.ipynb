# Ex. 9

```python
filename = '../datasets/Auto.csv'
data = pd.read_csv(filename, index_col='name')
data = data.loc[data.horsepower!='?']
data.horsepower = data.horsepower.astype(int)
data.head() 
```

```python
data.loc[data.isna().any(axis=1)]
```

```python
# Print count of different values
for col in data.columns:
    print(col, len(data[col].value_counts()))
```

```python
qual = ['cylinders', 'year', 'origin']
quant = ['mpg', 'displacement', 'horsepower', 'weight', 'acceleration']
```

```python
data.horsepower
```

```python
data.loc[:, quant].describe()
```

```python
data.loc[:, quant].drop(data.index[10:86], axis=0).describe()
```

```python
sns.pairplot(data)
```

```python
sns.heatmap(data.corr())
```

`mpg` predictor is almost completely negatively correlated with `cylinders`, `displacement`, `horsepower` and `weight` predictors. Also we can see a highly positive correlation between all four latter predictors. Other Correlation are not very strong, but also significant. Let's take a more close look at qualitative predictors.

```python
_, axes = plt.subplots(2,3, figsize=(15,12))
for i, col in enumerate(quant):
    sns.boxplot(data['cylinders'], data[col], ax=axes[i%2, i//2])
```

```python
_, axes = plt.subplots(2,3, figsize=(15,12))
for i, col in enumerate(quant):
    sns.boxplot(data['year'], data[col], ax=axes[i%2, i//2])
```

```python
_, axes = plt.subplots(2,3, figsize=(15,12))
for i, col in enumerate(quant):
    sns.boxplot(data['origin'], data[col], ax=axes[i%2, i//2])
```

`cylinders`, `displacement`, `horsepower` and `weight` predictors may be very useful for predicting `mpg` variable. But it is not very helpful to use them alll simultaneously, because they are very highly correlated between each other.
